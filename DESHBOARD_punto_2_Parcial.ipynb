```python
%%writefile THD.py
import streamlit as st
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as sig
import sympy as sym
st.sidebar.title("Presentado por:")
st.sidebar.write("Ever Daniel Hernández Tulcán")


# Título y descripción general
st.title("Análisis de un péndulo elástico amortiguado")

st.write("""
Las propiedades mecánicas del péndulo elástico amortiguado se analizan utilizando la teoría de los SLIT. La configuración mecánica subyacente se muestra en la siguiente figura:


![Péndulo Elástico Amortiguado](https://github.com/amalvarezme/SenalesSistemas/blob/master/4_Transformada_Laplace/damped_spring.png?raw=1)
""")


# Sección teórica
st.write("""
### Ecuación diferencial

La ecuación diferencia del sistema mecánico se obtiene al considerar el equilibrio de fuerzas ejercidas sobre la masa:
""")

st.latex(r"F_\text{S}(t) + F_\text{F}(t) + F_\text{I}(t) = F_\text{E}(t)")

st.write("""
Donde:

Donde $F_{E}(t) $ denota una fuerza externa que actúa sobre la masa. La fuerza $F_{S}(t)$ inducida por el resorte está dada por la ley de Hooke
""")
st.latex(r"F_\text{S}(t) = k y(t)")

st.write("""

Es común modelar la fuerza de fricción $F_{F}(t)$ inducida por el amortiguador de manera proporcional a la velocidad de la masa
""")
st.latex(r" F_\text{F}(t) = c \frac{d y(t)}{dt}")

st.write("""
La fuerza inercial $F_{I}(t)$ debida a la aceleración de la masa se da como
""")
st.latex(r" F_\text{I}(t) = m \frac{d^2 y(t)}{dt^2} ")

st.write("""
Al introducir las fuerzas anteriormente halladas en el equilibrio de fuerzas se obtiene la ecuación diferencial que describe el desplazamiento del péndulo elástico amortiguado
""")

st.latex(r"m\,\frac{d^{2}y(t)}{dt^{2}} + c\,\frac{dy(t)}{dt} + k\,y(t) = F_{E}(t)")


st.write("""
Como consecuencia de la fuerza externa.

La ecuación anterior constituye una EDO con coeficientes constantes. Puede interpretarse como un SLTI con la fuerza externa $x(t)= F_{E}(t)$ como señal de entrada y el desplazamiento de la masa como la señal de salida $y(t)$.
""")

# Inicializar impresión simbólica
sym.init_printing()

# Definir símbolos
t, m, c, k = sym.symbols('t m c k', real=True)
x = sym.Function('x')(t)
y = sym.Function('y')(t)

# Crear la ecuación diferencial
ode = sym.Eq(m * y.diff(t, 2) + c * y.diff(t) + k * y, x)

# Mostrar en pantalla
st.write("### Ecuación diferencial generada con SymPy:")
st.latex(sym.latex(ode))

st.write("""
Tomando transformada de Laplace (condiciones iniciales cero) y usando
$Y(s)$ Y $F(s)$ donde: """)


st.latex(r" H(s) = \frac{Y(s)}{F(s)}")

st.write("Función de transferencia")

st.latex(r"H(s)=\frac{1}{m s^2 + c s + k}")


# Título y descripción general
st.title("Análisis de un péndulo elástico amortiguado")

st.write("""
Las propiedades mecánicas del péndulo elástico amortiguado se analizan utilizando la teoría de los SLIT. La configuración mecánica subyacente se muestra en la siguiente figura:


![Péndulo Elástico Amortiguado](https://github.com/amalvarezme/SenalesSistemas/blob/master/Otros/circuitoRLC.jpeg?raw=2)
""")
st.write("Calculamos la Ecuación Diferencial")

st.write("La ecuación diferencial que describe la relación entrada/salida en el circuito se obtiene al aplicar las leyes de Kirchhoff. Así, se obtiene la siguiente EDO")

st.latex(r"il(t) = ic(t) + iR(t)")


st.latex(r"CL \frac{d^2 Vo(t)}{dt^2} + \frac{L}{R}\frac{d Vo(t)}{dt} + Vo(t) = Vi(t)")

st.write("Aplicando la trasfomada de laplace tenemos")

st.latex(r"CLs^2Vo(s)+ (LsVo(s)/R) + Vo(s) = Vi(s)")

st.write("""
Tomando transformada de Laplace (condiciones iniciales cero) y usando
$Y(s)$ Y $F(s)$ donde: """)


st.latex(r" H(s) = \frac{Y(s)}{F(s)}")

st.write("Función de transferencia")

st.latex(r"H(s)=\frac{1}{LC s^2 + \frac{Ls}{R} + 1}")

st.write("Comparación con la red eléctrica")

st.write("""
Comparando las EDO del péndulo elástico amortiguado y el circuito RLC, se tiene que ambos constituyen una EDO de segundo orden con coeficientes constantes.

La comparación de la EDO anterior con la del péndulo elástico amortiguado produce equivalencia entre sistemas

|                         | Circuito RLC | Péndulo elástico                          |
|-------------------------|------------------------------------|-------------------------------------------|
| Señal de entrada $x(t)$ | $V_{i}(t) = F_{E}(t) C$  |$F_{E}(t) = 1/C*V_{i}(t)$ |
| Señal de salida $y(t)$  | $V_{o}(t)$                    | $y(t)$                                    |
|                         | $L=m$                              | $m=L$                                     |
|                         | $R=L/Cc$                              | $c=L/RC$                                     |
|                         | $C= 1/k$                    | $k={1}/{C}$                           |
""")


#___________________________________________________________

import streamlit as st
import sympy as sym
import numpy as np
import matplotlib.pyplot as plt

# --- Configuración de impresión simbólica ---
sym.init_printing()

# --- Definición de variables simbólicas ---
t = sym.symbols('t', real=True)
m, c, k = sym.symbols('m c k', positive=True)
y = sym.Function('y')(t)

# --- Ecuación diferencial ---
x = sym.DiracDelta(t)
ode = sym.Eq(m * y.diff(t, 2) + c * y.diff(t) + k * y, x)

# -------------------------
# INTERFAZ DEL DASHBOARD
# -------------------------
st.title("Simulación de Sistema Sobreamortiguado")
st.write("Solución simbólica y simulación de la respuesta al impulso")

# Parámetros de usuario
m_val = st.number_input("Valor de m", value=1.0)
c_val = st.number_input("Valor de c (debe ser grande para sobreamortiguar)", value=3.0)
k_val = st.number_input("Valor de k", value=1.1)

mck = {m: m_val, c: c_val, k: k_val}

# -------------------------
# SOLUCIÓN DE LA ODE
# -------------------------
solution_h = sym.dsolve(
    ode.subs(y, sym.Function('h')(t))
)

# Encontrar constantes usando condiciones de continuidad
C1, C2 = sym.symbols("C1 C2")

eqs = [
    solution_h.rhs.limit(t, 0, "-"),
    solution_h.rhs.diff(t).limit(t, 0, "-")
]

constants = sym.solve(eqs, [C1, C2])
h = solution_h.subs(constants)

# -------------------------
# EVALUACIÓN NUMÉRICA
# -------------------------
h_lambda = sym.lambdify(t, h.rhs.subs(mck), "numpy")

T = np.linspace(0, 12, 1500)
H = h_lambda(T)

# -------------------------
# GRÁFICA
# -------------------------
st.subheader("Respuesta al impulso h(t) — Sistema Sobreamortiguado")

fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(T, H, label="h(t)", linewidth=2)
ax.set_xlabel("Tiempo [s]")
ax.set_ylabel("Amplitud")
ax.grid(True)
ax.legend()

st.pyplot(fig)


st.write("### Simulacion de sistema subamortiguado")
import streamlit as st
import sympy as sym
import numpy as np
import matplotlib.pyplot as plt

# -----------------------------------
# CONFIGURACIÓN SIMBÓLICA
# -----------------------------------
sym.init_printing()
t = sym.symbols("t", real=True)
m, c, k = sym.symbols("m c k", positive=True)
y = sym.Function("y")(t)

# Señal de entrada: impulso
x = sym.DiracDelta(t)
ode = sym.Eq(m * y.diff(t, 2) + c * y.diff(t) + k * y, x)

# -----------------------------------
# DASHBOARD
# -----------------------------------
st.title("Simulación de Sistemas de Segundo Orden")
st.write("Subamortiguado · Crítico · Sobreamortiguado")

# Parámetros del usuario
m_val = st.number_input("m", value=1.0)
c_val = st.number_input("c", value=0.5)
k_val = st.number_input("k", value=1.0)

# -----------------------------------
# CÁLCULOS DE PARÁMETROS
# -----------------------------------
omega_n = np.sqrt(k_val / m_val)
zeta = c_val / (2 * np.sqrt(m_val * k_val))

if zeta < 1:
    regimen = "Subamortiguado"
elif np.isclose(zeta, 1):
    regimen = "Amortiguamiento Crítico"
else:
    regimen = "Sobreamortiguado"


st.subheader("Clasificación del sistema")
st.write(f"**Factor de amortiguamiento:**  ζ = {zeta:.4f}")
st.write(f"**Frecuencia natural no amortiguada:**  ωₙ = {omega_n:.4f}")
st.write(f"**Régimen:**  **{regimen}**")

# Frecuencia amortiguada
if zeta < 1:
    omega_d = omega_n * np.sqrt(1 - zeta**2)
else:
    omega_d = None

# -----------------------------------
# TIEMPOS CARACTERÍSTICOS
# -----------------------------------
st.subheader("Tiempos característicos")

if zeta < 1:
    t_p = np.pi / omega_d
    t_r = 1.8 / omega_n
else:
    t_p = None
    t_r = None

t_s = 4 / (zeta * omega_n)

st.write(f"**Tiempo de pico (tₚ):** {t_p if t_p else 'No aplica'}")
st.write(f"**Tiempo de levantamiento (tᵣ):** {t_r if t_r else 'No aplica'}")
st.write(f"**Tiempo de establecimiento (tₛ):** {t_s:.4f}")

# -----------------------------------
# SOLUCIÓN SIMBÓLICA DE LA ODE
# -----------------------------------
solution_h = sym.dsolve(ode.subs(y, sym.Function("h")(t)))

C1, C2 = sym.symbols("C1 C2")

eqs = [
    solution_h.rhs.limit(t, 0, "-"),
    solution_h.rhs.diff(t).limit(t, 0, "-")
]

constants = sym.solve(eqs, [C1, C2])
h = solution_h.subs(constants)

h_fun = sym.lambdify(t, h.rhs.subs({m: m_val, c: c_val, k: k_val}), "numpy")

# -----------------------------------
# GRAFICACIÓN
# -----------------------------------
st.subheader("Respuesta al impulso h(t)")

T = np.linspace(0, 10, 2000)
H = h_fun(T)

fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(T, H, linewidth=2)
ax.set_xlabel("Tiempo [s]")
ax.set_ylabel("Amplitud")
ax.grid(True)

st.pyplot(fig)

```
